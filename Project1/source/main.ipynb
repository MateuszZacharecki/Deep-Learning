# CINIC-10 classification

This notebook presents results of the training.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

## CNN

To see the training process, please, check file /cnn/test.ipynb and /cnn/train_history for the results.

```python
cnn1_acc = pd.read_csv('cnn/train_history/cnn1/accuracy.csv')
cnn2_acc = pd.read_csv('cnn/train_history/cnn2/accuracy.csv')
cnn3_acc = pd.read_csv('cnn/train_history/cnn3/accuracy.csv')
cnn4_acc = pd.read_csv('cnn/train_history/cnn4/accuracy.csv')
cnn5_acc = pd.read_csv('cnn/train_history/cnn5/accuracy.csv')
cnn6_acc = pd.read_csv('cnn/train_history/cnn6/accuracy.csv')
cnn7_acc = pd.read_csv('cnn/train_history/cnn7/accuracy.csv')
cnn8_acc = pd.read_csv('cnn/train_history/cnn8/accuracy.csv')
cnn10_acc = pd.read_csv('cnn/train_history/cnn10/accuracy.csv')

accuracy_df = pd.DataFrame()
accuracy_df['CNN1'] = cnn1_acc['accuracy']
accuracy_df['CNN2'] = cnn2_acc['accuracy']
accuracy_df['CNN3'] = cnn3_acc['accuracy']
accuracy_df['CNN4'] = cnn4_acc['accuracy']
accuracy_df['CNN5'] = cnn5_acc['accuracy']
accuracy_df['CNN6'] = cnn6_acc['accuracy']
accuracy_df['CNN7'] = cnn7_acc['accuracy']
accuracy_df['CNN8'] = cnn8_acc['accuracy']
accuracy_df['CNN10'] = cnn10_acc['accuracy']

plt.boxplot(accuracy_df, labels=accuracy_df.columns)
plt.xlabel('Classifiers')
plt.ylabel('Accuracy')
plt.title('Accuracy boxplots for different architectures');
```

Each architecture training was tested 5 times. The following table contains mean and variance of accuracy of each architecture.

```python
mean_values = accuracy_df.mean(axis=0)
variance_values = accuracy_df.var(axis=0)
pd.DataFrame([mean_values, variance_values], index=['Mean', 'Variance'])
```

Learning rate tuning was done on the architecture with best accuracy â€“ CNN6 containing 6 convolutional layers and Adamax optimizer.

```python
cnn9_acc = pd.read_csv('cnn/train_history/cnn9/accuracy.csv')
cnn9_acc['accuracy']
l_rates = [0.0005, 0.001, 0.005, 0.01, 0.05, 0.1]

plt.plot(l_rates, cnn9_acc['accuracy'])
plt.title('Accuracy with different learning rate')
plt.xlabel('Learning rate')
plt.ylabel('Accuracy');
```

## VGG16

To see the training process, please, check file /vgg16/test.ipynb and /vgg16/train_history for the results.

```python
vgg1_acc = pd.read_csv('vgg16/train_history/vgg1/accuracy.csv')
vgg2_acc = pd.read_csv('vgg16/train_history/vgg2/accuracy.csv')
vgg3_acc = pd.read_csv('vgg16/train_history/vgg3/accuracy.csv')
vgg4_acc = pd.read_csv('vgg16/train_history/vgg4/accuracy.csv')

accuracy_df = pd.DataFrame()
accuracy_df['VGG1'] = vgg1_acc['accuracy']
accuracy_df['VGG2'] = vgg2_acc['accuracy']
accuracy_df['VGG3'] = vgg3_acc['accuracy']
accuracy_df['VGG4'] = vgg4_acc['accuracy']

plt.boxplot(accuracy_df, labels=accuracy_df.columns)
plt.xlabel('Classifiers')
plt.ylabel('Accuracy')
plt.title('Accuracy boxplots for different architectures');
```

First four architectures was tested 5 times. The following table contains mean and variance of accuracy of each architecture.

```python
mean_values = accuracy_df.mean(axis=0)
variance_values = accuracy_df.var(axis=0)
pd.DataFrame([mean_values, variance_values], index=['Mean', 'Variance'])
```

```python
vgg7_acc = pd.read_csv('vgg16/train_history/vgg7/accuracy.csv')
vgg8_acc = pd.read_csv('vgg16/train_history/vgg8/accuracy.csv')
vgg9_acc = pd.read_csv('vgg16/train_history/vgg9/accuracy.csv')
rate = [0.0005, 0.001, 0.005, 0.01, 0.05]
batch = [32, 64, 128, 256, 512]
```

Accuracies for VGG7, VGG8 and VGG9 tests are presented in the following three graphs, depending on tuned parameters. 

```python
plt.plot(rate, vgg7_acc['accuracy'])
plt.title('Accuracy with different learning rate')
plt.xlabel('Learning rate')
plt.ylabel('Accuracy');
```

```python
plt.plot(rate, vgg8_acc['accuracy'])
plt.title('Accuracy with different learning rate')
plt.xlabel('Learning rate')
plt.ylabel('Accuracy');
```

```python
plt.plot(batch, vgg9_acc['accuracy'])
plt.title('Accuracy with different batch size')
plt.xlabel('Batch size')
plt.ylabel('Accuracy');
```

The best trained VGG model turned out to be VGG16 model with normalization layer and denses with respectively 50, 20 and 10 units, with adamax optimizer and learning rate = 0.001, trained with batch size = 128 (VGG9 with batch size = 128). This model achieved accuracy 54.34%.The following graphs present respectively accuracy per epoch, loss per epoch and confusion matrix for this model.
